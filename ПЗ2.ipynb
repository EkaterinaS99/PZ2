# *Исследование режимов работы линии передачи*  
Получить эпюры тока и напряжения, соответствующие всем режимам работы длинной линии.
Для этого смоделировать работу линии на следующие виды нагрузок (Zн = Rн + jXн):  

1. Линия, нагруженная на реактивное сопротивление (Zн = jXн)  
2. Линия, нагруженная активное сопротивление (Zн = Rн , рассмотреть случаи: Rн > W; Rн < W; Rн = W)  
3. Линия, нагруженная на комплексное сопротивление (Zн = Rн + jXн)  

```python
import numpy as np
import matplotlib.pyplot as plt
```

### *Линия, нагруженная на реактивное сопротивление:*

```python
W = 1
lmd = 1
Un=1 
```

```python
Xn1=1*W
N=1000
z=np.zeros(N)
dz=2/(N-1)*lmd
for i in range(N):
    z[i]=dz*i
Zn1=1j*Xn1
I1=Un/Zn1*np.cos(2*np.pi/lmd*z)+Un/W*np.sin(2*np.pi/lmd*z)*1j
U1=Un*np.cos(2*np.pi/lmd*z)+Un/Zn1*W*np.sin(2*np.pi/lmd*z)*1j
```

```python
fig, ax = plt.subplots(2,1,figsize=(10, 10), dpi=600)
ax[0].plot(z,np.abs(U1))
ax[0].grid() #сетка
ax[0].set_xlim([2,0]) #пределы по оси х
ax[0].set_ylim([0, 1.5*max(abs(U1))]) #пределы по оси y
ax[0].set_xlabel('$z/\lambda$')
ax[0].set_ylabel('|U|, В')
ax[0].set_title("Распределение амплитуд напряжения")
ax[0].plot(z,np.abs(U1),'r')
ax[1].grid()
ax[1].set_xlim([2,0])
ax[1].set_title("Распределение амплитуд тока")
ax[1].plot(z,np.abs(I1),'darkorange')
ax[1].set_xlabel("$z/\lambda$")
ax[1].set_ylabel('|I|, А')
ax[1].set_title("Распределение амплитуд тока")
```

Расчитаем КбВ и коэффициент отражения

```python
G1=((Zn1-W)/(Zn1+W))
Kbv1=(1-abs(G1))/(1+abs(G1))
print("Кбв =",Kbv1,
     "|Г| =",abs(G1))
```

```python
# активное сопротивление Rn=W
Rn2=1*W
N=1000
Zn2=Rn2
I2=Un/Zn2*np.cos(2*np.pi/lmd*z)+Un/W*np.sin(2*np.pi/lmd*z)*1j
U2=Un*np.cos(2*np.pi/lmd*z)+Un/Zn2*W*np.sin(2*np.pi/lmd*z)*1j

```

```python
G2=((Zn2-W)/(Zn2+W))
Kbv2=(1-abs(G2))/(1+abs(G2))
print("Кбв =",Kbv2,
     "|Г| =",abs(G2))
```
